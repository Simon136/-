```python
#Timestamp类型 用to_datetime进行转换
data['日期'] = pd.to_datetime(data['日期'])
```

```python
#城市 分类
city = data.groupby(by = '城市')
# city.count()
```

```python
#设置 新建列名
data['累计确诊人数'] = 0
data['累计治愈人数'] = 0
data['累计死亡人数'] = 0
```

```python
#通过分类和cumsum函数实现累加
data['累计确诊人数'] = city['新增确诊'].apply(lambda x:x.cumsum())
data['累计治愈人数'] = city['新增治愈'].apply(lambda x:x.cumsum())
data['累计死亡人数'] = city['新增死亡'].apply(lambda x:x.cumsum())
data.head()
```

```python
#保存数据
data1 = data[['城市','日期','累计确诊人数','累计治愈人数','累计死亡人数']]
data1.to_csv(r'C:\Users\Radish\Desktop\data B\task1_1.csv',index=False)
```

```python
# 建立透视表
pd.pivot_table(data,index=['城市','日期','累计确诊人数','累计治愈人数','累计死亡人数'])
```

```python
#定义一个函数 作为返回10、25号的累计数据
def sum(city):
    #每月10日的数据
    mon_10 = city.loc[city.日 == 10]

    #每月25日的数据
    mon_25 = city.loc[city.日 == 25]
    
    #返回值
    return mon_10,mon_25
```

```python
#使用匹配机制让两个表相关
df1 = pd.read_csv(r'D:\大数据学习\数据分析与挖掘\20B题\数据\task1_1.csv')
df2=pd.read_excel(r'D:\大数据学习\数据分析与挖掘\20B题\附件1.xlsx','城市省份对照表')

#用merge方法关联两个表
outer=pd.merge(df1,df2,on='城市')

#按省份分类
city = outer.groupby(by='省份')
data1 = outer[['省份','日期','新增确诊','新增治愈','新增死亡','累计确诊人数','累计治愈人数','累计死亡人数']]

# 导出数据
data1.to_csv(r'D:\大数据学习\数据分析与挖掘\20B题\数据\task1_2.csv',index=False)
```

```python
#导入数据
data1_2=pd.read_csv(r"D:\大数据学习\数据分析与挖掘\20B题\数据\task1_2.csv")

#查看表结构
data1_2.dtypes
```

```python
#转化时间格式
data1_2.日期=pd.to_datetime(data1_2.日期,format='%Y/%m/%d')
```

```python
#提取 日、月的数据
data1_2['月']=data1_2['日期'].dt.month
data1_2['日']=data1_2['日期'].dt.day

#查看表
data1_2
```

```python
#提取省份为湖北的每月15号的数据
data1_2_1=data1_2.loc[data1_2['省份']=='湖北']
data1_2_2=data1_2_1.loc[data1_2_1['日']==15]

#查看表
data1_2_2
```

```python
#提取省份为广东的每月15号的数据
data1_2_3=data1_2.loc[data1_2['省份']=='广东']
data1_2_4=data1_2_3.loc[data1_2_3['日']==15]

#查看表
data1_2_4
```

```python
##提取省份为河北的每月15号的数据
data1_2_5=data1_2.loc[data1_2['省份']=='河北']
data1_2_6=data1_2_5.loc[data1_2_5['日']==15]

#查看表
data1_2_6
```

```python
#链接三个分析表 形成透视图 更加直观
data1_2_t=pd.concat([data1_2_2,data1_2_4,data1_2_6])
data1_2_t1=data1_2_t.groupby(by=['省份','日期'])   #按省份和日期分类
data1_2_t2=data1_2_t1.aggregate({'累计确诊人数':np.sum,'累计治愈人数':np.sum,'累计死亡人数':np.sum})  #通过np算出数据
data1_2_t2
```
